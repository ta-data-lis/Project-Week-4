```python
import pandas as pd
import numpy as np
import re
from scipy.stats import ttest_1samp, ttest_ind
import scipy.stats as st
#pd.set_option('max_rows', 30)
```

```python
file1= pd.read_csv('HAVE_RWB.csv') #Read
file1.head()
```

```python
#DROPING the 'Timestamp' column that comes from the Google form
file1=file1.drop('Timestamp', axis=1)
```

```python
file1.reset_index(drop=True) #index reset following the drop
file1.head()
```

```python
#RENAMING the column names from file1 - renaming all columns method
file1.columns=['Why_buy','Own_plus','Material','Env_impact','Env_incentive','YN_technical','Detail_technical','YN_disadvantage','Detail_disadvantage','Design_VS_Pract','Price_Range','Bought_Where','Use']
file1.head()
```

```python
def tonumb (row): # function that change the string "Yes" into 1 ("No" becomes 0) ##This is to do the hypothesis testing 
    if row =='Yes':
        return 1
    else:
        return 0
```

```python
file1['Own_plus']= file1['Own_plus'].apply(tonumb) #applying the function to the row in the column
```

```python
file1['Env_incentive']= file1['Env_incentive'].apply(tonumb) #applying the function to the row in the column
```

```python
###########CLEANING THE DATA####################

#1) Uniformizing the Yes answers (whole dataframe)
file1=file1.replace(to_replace ="Yes, explain below", value ="Yes")
```

```python
#2)Removing some unwanted punctuation
file1['Design_VS_Pract']=file1['Design_VS_Pract'].str.replace("Both.","Both")
```

```python
file1['Design_VS_Pract']=file1['Design_VS_Pract'].str.replace("/W","")
file1.head()
```

```python
#######TESTING the hypothesis' sur Database 1 ####################
```

```python
elseglass=file1[file1['Material']!='Glass']
elseglass
```

```python
glassf=file1[file1['Material']=='Glass'] # Selecting and adding to new df Glass rows.
glassf
```

```python
#H1 --- Are the people who buy glass bottle concerned by the environment incentive? 
#H0 ---Does having environmental concerns equates for individuals to possess a glass bottle?
st.ttest_ind(glassf['Env_incentive'],elseglass['Env_incentive'], equal_var=False)
#Answer : In this sample, we cannot reject the hypothesis
```

```python
#H0---  Is the fact that people have a glass bottle, related to having more than one bottle?
#H1---Do people who have a glass bottle own more than one reusable water?
st.ttest_ind(glassf['Own_plus'], elseglass['Own_plus'], equal_var=False)
#having more bottles is not different from glass and non glass bottle owners #we cannot reject the H0 on this sample
```
