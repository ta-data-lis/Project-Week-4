# Rare earth Datasource

```python
import pandas as pd
ree = pd.read_csv("Global_REE_occurrence_database.csv", sep=';')
ree.columns
```

#### Clean Data with all columns with more than 2000 NaN

```python
ree_isnull = ree.isnull().sum()
ree_cols = ree_isnull[ree_isnull > 2000].index
ree_drop = ree.drop(ree_cols, axis = 1)
set(ree_drop["P_Status"])
```

```python
# unify the fillings in column P_Status

ree_drop["P_Status"]= ["Byproduct producer" if (ree_drop["P_Status"][i] == 'Byproduct producer(?)') | (ree_drop["P_Status"][i] == 'Byproduct producer') 
                       else "No Production" if (ree_drop["P_Status"][i] == 'No produciton(?)') | (ree_drop["P_Status"][i] == 'No production') |  (ree_drop["P_Status"][i] == 'No production(?)')
                       else "Past byproduct producer" if (ree_drop["P_Status"][i] == 'Past byproduct producer') | (ree_drop["P_Status"][i] == 'Past byproduct producer(?)')
                       else 'Past producer' if (ree_drop["P_Status"][i] == 'Past producer') | (ree_drop["P_Status"][i] == 'Past producer(?)')
                       else "Producer" if (ree_drop["P_Status"][i] == 'Producer') | (ree_drop["P_Status"][i] == 'Producer(?)')
                       else "Not known" for i in range(len(ree_drop["P_Status"]))]
```

```python
ree = ree_drop
ree
```

### Producer / Mining of REE
