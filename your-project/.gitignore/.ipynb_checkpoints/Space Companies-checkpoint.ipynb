```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Data via Web-Scraping

```python
url = "https://en.wikipedia.org/wiki/List_of_private_spaceflight_companies#Satellite_launchers"
```

```python
html = requests.get(url).content

soup = BeautifulSoup(html, "lxml")  
soup
```

### Privat Mining Companies

```python
data = [i.text.replace("\n\n",";") for i in soup.find_all("table")]
mining = data[10].split("\n")
mining = [mining[i].split(";") for i in range(10)]
```

```python
ds_mining = pd.DataFrame(mining)
ds_mining = ds_mining[(ds_mining[0]!="")]
ds_mining = ds_mining.rename(columns={0:"Company name",1:"Body to be mined", 2:"Mining craft", 3:"Mining status",4:"Ref",5:"Nation"})
ds_mining
```

```python
mining_nationality = ["United States", "-", "Japan", "United States", "United States","United States","India", "United States"]
ds_mining["Nation"] = mining_nationality
ds_mining
```

### Privat Launch Vehicle makers

```python
launcher = data[4].split("\n")
launcher = [launcher[i].split(";") for i in range(84)]
```

```python
ds_launcher = pd.DataFrame(launcher)
ds_launcher = ds_launcher[(ds_launcher[0]!="")&
                         (ds_launcher[3] == "1")|
                         (ds_launcher[3] == "2")|
                         (ds_launcher[3] == "3-4")|
                         (ds_launcher[3] == "1 + airplane")|
                         (ds_launcher[3] == "variable")|
                         (ds_launcher[3] == "2 + 0-5 boosters")|
                         (ds_launcher[3] == "2 + airplane")|
                         (ds_launcher[3] == "3 + high-altitude balloon")]

ds_launcher= ds_launcher.rename(columns={0:"Company name",1:"Launcher name",2:"Launcher type",3:"No. of stages",4:"Maximum reach",5:"Launcher status",6:"Maiden flight",7:"Ref",8:"Nation"})
launcher_nation = ["United States","Australia","United States","United States","Canada","-","United States","United States","Australia","Malaysia","United States","Japan", "Italy","United States","United States","China","United Kingdom","Germany","Spain","United States","United States","United States","United States","Scotland","Poland","United States","-","United States","United States","United States","United States","Spain"]

ds_launcher["Nation"]=launcher_nation
ds_launcher = ds_launcher[(ds_launcher["Company name"] != "Canadian Arrow")&
                          (ds_launcher["Company name"] != "Lockheed Martin")&
                          (ds_launcher["Company name"] != "Orbital Transport & Raketen AG")&
                          (ds_launcher["Company name"] != "Space Services Inc.")&
                          (ds_launcher["Company name"] != "Datiotec Aeroespacial / INMEU A.C.")&  
                          (ds_launcher["Company name"] != "SpaceLS")]
(ds_launcher)
```

```python
launcher_per_nation = ds_launcher.groupby("Nation", as_index = False).agg({"Company name":"count"}).sort_values(by = "Company name", ascending = False)
launcher_per_nation
```

# Government space agencies Budget

```python
url = "https://en.wikipedia.org/wiki/List_of_government_space_agencies"
```

```python
html = requests.get(url).content

soup = BeautifulSoup(html, "lxml")  
soup
```

```python
budget = [i.text.replace("\n\n",";").replace("\xa0", "") for i in soup.find_all("table")]
budget = budget[6].split("\n")
budget = [budget[i].split(";") for i in range(32)]
```

```python
ds_budget = pd.DataFrame(budget)
ds_budget = ds_budget[(ds_budget[0]!="")]
ds_budget = ds_budget.rename(columns = {0:"Country",1:"Agency",2:"Budget",3:"Ref",4:"Year"})
```

```python
ds_budget["Country"] = ["European Union" if ds_budget["Country"][i] == "AustriaBelgiumCzech RepublicDenmarkEuropean UnionFinlandFranceGermanyGreeceIrelandItalyLuxembourgNetherlandsNorwayPolandPortugalRomaniaSpainSwedenSwitzerlandUnited Kingdom" else ds_budget["Country"][i] for i in range(1,32)]
ds_budget.sort_values(by = "Budget", ascending = True)
ds_budget["Budget"] = [ds_budget.Budget[i].replace(",","") for i in range(1,32)]
ds_budget["Budget"] = ds_budget.Budget.astype(float)
ds_budget
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from scipy import stats
```

```python
sns.set_style("whitegrid")
ax = sns.barplot(x = "Country", y = "Budget",  data = ds_budget)
ax.set_title("Budget per Country")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)

ax.figure.savefig("Budget_per_Country.jpg")
```

### Import REE per country Data to analyse the Correlation between Budget and Occurence per Country

```python
ree = pd.read_csv("ree_per_country.csv")
ree
```

```python
correlation = ree.merge(ds_budget, on = "Country")
```

```python
correlation.corr()
```

```python
sns.set_style("whitegrid")
ax = sns.scatterplot( x = "Commods",y = "Budget" ,data = correlation,  hue = "Country", s = 500, alpha = 0.8)

ax.text(214+15, 8400, "China", horizontalalignment='left', size='medium', color='black')
ax.text(153+15, 21500, "United States", horizontalalignment='left', size='medium', color='black')
ax.text(466+15, 6.9, "Australia", horizontalalignment='left', size='medium', color='black')
ax.text(254+15, 246, "Canada", horizontalalignment='left', size='medium', color='black')

ax.set_title("Occurrence vs. Budget per Country",y= 1.1, fontsize=15, weight = "semibold")
ax.set_xlabel("Occurrence of REE", fontsize = 13, weight = "semibold")
ax.set_ylabel("Budget in Mio. USD $", fontsize = 13, weight = "semibold")

ax.legend(loc='center right', bbox_to_anchor=(1.5, 0.5), ncol=1)


x = correlation['Commods']
y = correlation['Budget']
slope, intercept, r_value, _, std_err = stats.linregress(x, y)

ax.plot(x, slope*x + intercept, color="red")
plt.show()


ax.figure.savefig("Occurrence_vs_Budget.jpg")
```
