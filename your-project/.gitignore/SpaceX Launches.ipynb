```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = "https://api.spacexdata.com/v3/launches"
launches = requests.get(url)#.json()

launches
```

```python
ds = pd.DataFrame(launches)
ds["rocket"]
```

```python
norm_json = json_normalize(launches.json())
ds_normalized = pd.DataFrame(norm_json)
rocket = ds_normalized["rocket.second_stage.payloads"]
```

```python
df = pd.DataFrame(rocket)
```
